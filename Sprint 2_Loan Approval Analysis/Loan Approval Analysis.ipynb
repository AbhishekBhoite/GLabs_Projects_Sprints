```python
#Drop the column Loan_ID and make new dataframe banks
banks=bank.drop(["Loan_ID"], axis=1)
banks.head()
```

```python
#Number of rows in each column with NaN values
banks.isnull().sum()
```

```python
#Print all rows with one or more NaN values 
banks[banks.isnull().any(axis=1)]
```

```python
#STEP 2: Get columns that have missing values, try to fill them.
for column in banks[['Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area','Loan_Status']]:
    bank_mode = banks[column].mode()
    print(bank_mode)
    banks[column].fillna(bank_mode[0], inplace=True)
print(banks.isnull().sum())
```

```python
#Check if any rows are left with one or more NaN values 
banks[banks.isnull().any(axis=1)]
```

```python
#STEP 3: Check the loan amount of an average person
#print(banks.head())
avg_loan_amount=pd.pivot_table(banks, index=['Gender', 'Married', 'Self_Employed'], values='LoanAmount',aggfunc=['mean'])
print(avg_loan_amount)
```

```python
#STEP 4: Check the percentage of loan approved based on a person's employment type.
banks1= banks[ banks['Self_Employed']=='Yes']
#type(banks1['Loan_Status'].values[0])
banks2= banks1[ banks1['Loan_Status'] == 'Y']
#print(banks2.head())
loan_approved_se= len(banks2.index)
print(loan_approved_se)

banks3= banks[ banks['Self_Employed']=='No']
#type(banks1['Loan_Status'].values[0])
banks4= banks3[ banks3['Loan_Status'] == 'Y']
#print(banks4.head())
loan_approved_nse= len(banks4.index)
print(loan_approved_nse)
Loan_Status=614
percentage_se=loan_approved_se/Loan_Status*100
print(percentage_se)
percentage_nse=loan_approved_nse/Loan_Status*100
print(percentage_nse)
```

```python
#STEP 5: Find out those applicants with long loan amount term.
loan_term= banks['Loan_Amount_Term'].apply(lambda x: x/12) 
print(loan_term)
big_loan_term=len(loan_term[loan_term.values >= 25])
print(big_loan_term)
```
