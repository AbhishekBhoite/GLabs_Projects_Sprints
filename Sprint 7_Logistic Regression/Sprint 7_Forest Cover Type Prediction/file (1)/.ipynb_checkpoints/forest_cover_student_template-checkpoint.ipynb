### 1. Split the train_data into train and validation set. Standardize and then apply Logistic Regression Model

```python
# Split the train data into train and validation set
X= train_data.drop('Cover_Type',axis=1)
y= train_data['Cover_Type']
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
# Standardize data
scaler= StandardScaler()
scaler.fit(X_train)
X_train= scaler.transform(X_train)
X_test= scaler.transform(X_test)

# fit the model
model= LogisticRegression()
model.fit(X_train,y_train)
```

```python
#Finding the model score
accuracy_score = model.score(X_test, y_test)
print("Model Score after standardizing data is:",accuracy_score)
```

### 2. Analysis of variance (ANNOVA)

```python
#no of features list
nof_list=[10,15,20,25,30,35,40,45,50]

#Variable to store the highest score
high_score=0
#Variable to store the optimum features
nof=0
```

```python
from sklearn.feature_selection import f_regression
```

```python
# Split the train data into train and validation set
X= train_data.drop('Cover_Type',axis=1)
y= train_data['Cover_Type']
for n in nof_list:
    X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42,test_size=0.2)
    test= SelectKBest(score_func=f_regression,k='all')
    #Fitting and transforming the model on X_train
    X_train= test.fit_transform(X_train,y_train)
    #Fitting and transforming the model on X_test
    X_test=test.transform(X_test)
    #Initialising the Linear Regression model
    model= LogisticRegression()
    #Fitting the model
    model.fit(X_train,y_train)
    #Finding the model score
    print("For no of features=",n,", score=", model.score(X_test,y_test))
    if model.score(X_test,y_test)>high_score:
        high_score=model.score(X_test,y_test)
        nof=n
print("High Score is:",high_score, "with features=",nof)
```

### 3. Recursive Feature Elimination(RFE)

```python
# Split the train data into train and validation set
X= train_data.drop('Cover_Type',axis=1)
y= train_data['Cover_Type']
```

```python
#no of features list
nof_list=[10,15,20,25,30,35,40,45,50]

#Variable to store the highest score
high_score=0
#Variable to store the optimum features
nof=0
```

```python
for n in nof_list:
    X_train,X_test,y_train,y_test= train_test_split(X,y,random_state=42,test_size=0.2)
    model= LogisticRegression()
    rfe= RFE(model,n)
    X_train_rfe= rfe.fit_transform(X_train,y_train)
    X_test_rfe= rfe.transform(X_test)
    model.fit(X_train_rfe,y_train)
    R_2_score= model.score(X_test_rfe, y_test)
    print("For no of features=",n,", score=", R_2_score)
    if R_2_score>high_score:
        high_score= R_2_score
        nof= n
print("Highest Score from RFE Method is:",high_score," with No of Features-",nof)
```

### 4. Principal Component Analysis(PCA)

```python
from sklearn.decomposition import PCA
```

```python
#no of features list
nof_list=[10,15,20,25,30,35,40,45,50]

#Variable to store the highest score
high_score=0
#Variable to store the optimum features
nof=0
```

```python

X= train_data.drop('Cover_Type',axis=1)
y= train_data['Cover_Type']
```

```python
#Initialising PCA
for n in nof_list:
    pca= PCA(n_components=25,random_state=0)
    # Split the train data into train and validation set
    X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42,test_size=0.2)
    # fit the model
    model= LogisticRegression()
    model.fit(X_train,y_train)
    #Transforming the features
    X_train_pca= pca.fit_transform(X_train)
    X_test_pca= pca.transform(X_test)
    #Initialising the model
    model= LogisticRegression()
    #Fitting the model
    model.fit(X_train_pca,y_train)
    #Scoring the model
    pca_score= model.score(X_test_pca,y_test)
    print("For no of features=",n,", score=", pca_score)
    if pca_score>high_score:
        high_score= pca_score
        nof= n
print("Highest Score from PCA Method is:",high_score," with No of Features-",nof)
```
