```python
# to apply transformation to any column

def power_trans(df,col,power):
    lst = []
    for i in df[col]:
        if i < 0:
            lst.append(-np.power(abs(i),power))
        else:
            lst.append(np.power(i,power))
    return pd.DataFrame(lst)
```

```python
# Apply some good transformation on continuous features. Call transform_column() to do the same.
train['Age']= power_trans(train,'Age',0.05)
train['Overall']= power_trans(train,'Overall',0.5)
train['Potential']= power_trans(train,'Potential',1.1)
train['Wage (M)']= power_trans(train,'Wage (M)',0.15)
```

```python
# Before we proceed with the model fitting, we need to get rid of the categorical columns.
# We can use One Hot Encoding or Label Encoding to convert the categorical columns to numeric.
#Call the encode_category() with the list of columns and the method of encoding to do the same.
train= pd.concat([pd.get_dummies(train['Position']),train], axis=1)
```

```python
# Encoding the categorical columns dropped the accuracy so instead drop the columns  

train.drop(['Nationality','Position'],axis=1, inplace=True)
print(train.shape)
train.head()
```

### Model building

- Separate the features and target and then split the train data into train and validation set.
- Now let's come to the actual task, using linear regression, predict the `Value (M)`. 
- Try improving upon the `r2_score` (R-Square) using different parameters that give the best score. You can use higher degree [Polynomial Features of sklearn](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html) to improve the model prediction. 



```python
# Code Starts here
# Split the train data into train and validation set
X= train.drop('Value (M)',1)
y= train['Value (M)']
highest_r2_scr= 0
chosen_state= 0

for n in range(10):
    random_state=n
    X_train,X_test,y_train,y_test= train_test_split(X,y,random_state=random_state,test_size=0.3)
    
    #Standardize the data
    Scaler= StandardScaler()
    X_train= Scaler.fit_transform(X_train)
    X_test= Scaler.transform(X_test)
    
    #Initialize the Linear Regression Model
    model=Ridge()
    
     #Polynomial Features
    poly= PolynomialFeatures()
    X_train=poly.fit_transform(X_train)
    X_test=poly.transform(X_test)
    
    #Predict the accuracy score
    model.fit(X_train,y_train)
    pred=model.predict(X_test)
    r2_scr= r2_score(y_test,pred)
    print(r2_scr)
    if r2_scr>highest_r2_scr:
        highest_r2_scr=r2_scr
        chosen_state=n
print("The highest r2_score using Linear Regression Model is {} with chosen random state {}".format(highest_r2_scr,chosen_state))
# Code ends here.
```
